# Задание 1

```python
import pandas as pd
from sklearn.model_selection import train_test_split
from sklearn.linear_model import LinearRegression
```

```python
df = pd.read_csv('BostonHousing.csv', quotechar='"')
```

```python
X = df.drop('medv', axis=1)
y = df['medv']
```

```python
X_train, X_test, y_train, y_test = train_test_split(X, y, test_size=0.3, random_state=42)
```

```python
lr = LinearRegression()
```

```python
lr.fit(X_train, y_train)
```

```python
y_pred = lr.predict(X_test)
```

# Задание 2

```python
from sklearn.ensemble import RandomForestRegressor
```

```python
model = RandomForestRegressor(n_estimators=1000, max_depth=12, random_state=42)
```

```python
model.fit(X_train, y_train)
```

```python
y_pred_rf = model.predict(X_test)
```

```python
from sklearn.metrics import r2_score
r2 = r2_score(y_test, y_pred_rf)
```

```python
print(f'R2 для модели LinearRegression: {r2_score(y_test, y_pred)}')
print(f'R2 для модели RandomForestRegressor: {r2}')
```

# *Задание 4

```python
from sklearn.ensemble import RandomForestClassifier
from sklearn.model_selection import GridSearchCV
from sklearn.metrics import roc_auc_score
```

```python
df = pd.read_csv('creditcard.csv')
```

```python
print(df['Class'].value_counts(normalize=True))
```

```python
print(df.info())
```

```python
pd.options.display.max_columns = 100
```

```python
print(df.head(10))
```

```python
X = df.drop('Class', axis=1)
y = df['Class']
```

```python
X_train, X_test, y_train, y_test = train_test_split(X, y, test_size=0.3, random_state=100, stratify=y)
```

```python
print(X_train.shape, X_test.shape, y_train.shape, y_test.shape)
```

```python
parameters = [{'n_estimators': [10, 15], 'max_features': np.arange(3, 5), 'max_depth': np.arange(4, 7)}]
clf = GridSearchCV(estimator=RandomForestClassifier(random_state=100), param_grid=parameters, scoring='roc_auc', cv=3)
```

```python
clf.fit(X_train, y_train)
```

```python
print(clf.best_params_)
```

```python
clf = RandomForestClassifier(max_depth=6, max_features=3, n_estimators=15)
clf.fit(X_train, y_train)
```

```python
y_pred_proba = clf.predict_proba(X_test)[:, 1]
```

```python
auc = roc_auc_score(y_test, y_pred_proba)
print(f'AUC на тестовых данных: {auc}')
```

#### 9. Создайте список high_corr из признаков, корреляция которых с полем target по абсолютному значению превышает 0.5 (причем, само поле target не должно входить в этот список).

```python
high_corr = [col for col in X_corr.columns if abs(X_corr[col]['target']) > 0.5 and col != 'target']
high_corr
```

Выведите описание полей датафрейма X с помощью метода describe.

```python
X.drop('target', axis=1, inplace=True)
```
